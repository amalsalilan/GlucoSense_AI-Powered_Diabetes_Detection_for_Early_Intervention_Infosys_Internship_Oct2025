```python
plt.figure(figsize=(12, 6))
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['bmi'], bins=30, kde=True)
plt.title("BMI Distribution")
plt.show()
```

```python
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['bmi_category'] = df['bmi'].apply(categorize_bmi)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='diabetes', y='age', data=df)
plt.title("Age vs. Diabetes")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='diabetes', y='bmi', data=df)
plt.title("BMI vs. Diabetes")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=['number'])  # Select only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
comorbid_counts = df.groupby(['hypertension', 'heart_disease'])['diabetes'].mean().unstack()
comorbid_counts.plot(kind='bar', figsize=(10, 5), stacked=True)
plt.title("Diabetes Prevalence by Hypertension & Heart Disease")
plt.show()
```

```python
gender_counts = df.groupby('gender')['diabetes'].mean()
gender_counts.plot(kind='bar', color=['blue', 'pink', 'gray'], figsize=(8, 5))
plt.title("Diabetes Prevalence by Gender")
plt.ylabel("Proportion with Diabetes")
plt.show()
```
