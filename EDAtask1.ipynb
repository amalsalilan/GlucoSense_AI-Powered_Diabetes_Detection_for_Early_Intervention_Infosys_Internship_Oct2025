```python
import seaborn as sns
import matplotlib.pyplot as plt
num_cols = ['Glucose_Level', 'Insulin_Level', 'BMI', 'Age', 'HbA1c']
df[num_cols].hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x="Glucose_Level", y="Insulin_Level", hue="Diabetes_Status", palette="coolwarm")
plt.title("Glucose vs Insulin Levels by Diabetes Status")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df["Diabetes_Status"] == "Diabetic"]["Glucose_Level"], fill=True, color="red", label="Diabetic")
sns.kdeplot(df[df["Diabetes_Status"] == "Non-Diabetic"]["Glucose_Level"], fill=True, color="green", label="Non-Diabetic")
plt.title("Glucose Level Distribution by Diabetes Status")
plt.legend()
plt.show()
```
