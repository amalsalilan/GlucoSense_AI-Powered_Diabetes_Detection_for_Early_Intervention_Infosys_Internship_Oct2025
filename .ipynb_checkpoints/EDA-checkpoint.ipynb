# 1. Data Quality and Integrity Check

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Kaggle path)
file_path = 'C:/Users/Rishitha.G/Downloads/diabetes_prediction_dataset/diabetes_prediction_dataset.csv'
df = pd.read_csv(file_path)

# Display the first few rows
display(df.head())


```

```python
# Detect outliers using boxplots for BMI, Blood Glucose Level, and HbA1c Level
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(y=df['bmi'], ax=axes[0])
axes[0].set_title('BMI Outliers')
sns.boxplot(y=df['blood_glucose_level'], ax=axes[1])
axes[1].set_title('Blood Glucose Level Outliers')
sns.boxplot(y=df['HbA1c_level'], ax=axes[2])
axes[2].set_title('HbA1c Level Outliers')
plt.show()
```

```python
# Distribution of gender and smoking history
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.countplot(x='gender', data=df, ax=axes[0])
axes[0].set_title('Gender Distribution')
sns.countplot(x='smoking_history', data=df, ax=axes[1])
axes[1].set_title('Smoking History Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Class distribution of the target variable (diabetes)
sns.countplot(x='diabetes', data=df)
plt.title('Diabetes Class Distribution')
plt.show()
```

# 2. Univariate Analysis

```python
# Visualize distributions
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.histplot(df['age'], kde=True, ax=axes[0, 0]).set_title('Age Distribution')
sns.histplot(df['bmi'], kde=True, ax=axes[0, 1]).set_title('BMI Distribution')
sns.histplot(df['HbA1c_level'], kde=True, ax=axes[1, 0]).set_title('HbA1c Level Distribution')
sns.histplot(df['blood_glucose_level'], kde=True, ax=axes[1, 1]).set_title('Blood Glucose Level Distribution')
plt.tight_layout()
plt.show()

# Categorize BMI
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['bmi_category'] = df['bmi'].apply(categorize_bmi)
sns.countplot(x='bmi_category', data=df, order=['Underweight', 'Normal', 'Overweight', 'Obese'])
plt.title('BMI Categories')
plt.show()
```

# 3. Bivariate and Multivariate Analysis

```python
# Age vs. Diabetes and BMI vs. Diabetes
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(x='diabetes', y='age', data=df, ax=axes[0]).set_title('Age vs. Diabetes')
sns.boxplot(x='diabetes', y='bmi', data=df, ax=axes[1]).set_title('BMI vs. Diabetes')
plt.show()

# Impact of Smoking History on Diabetes Prevalence
sns.countplot(x='smoking_history', hue='diabetes', data=df)
plt.title('Impact of Smoking History on Diabetes')
plt.xticks(rotation=45)
plt.show()

# Correlation heatmap for numeric features only
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Generate the correlation matrix
plt.figure(figsize=(10, 8))
correlation = numeric_df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```

# 4. Comorbidity Analysis
- Analyzing the co-occurrence of hypertension and heart disease with diabetes.
- Evaluating how the number of comorbidities affects diabetes risk.

```python
# Create a new feature for the number of comorbidities
df['comorbidity_count'] = df['hypertension'] + df['heart_disease']

# Visualize the relationship between comorbidities and diabetes
import seaborn as sns
import matplotlib.pyplot as plt

# Count plot for diabetes occurrence by number of comorbidities
sns.countplot(x='comorbidity_count', hue='diabetes', data=df)
plt.title('Diabetes Prevalence by Number of Comorbidities')
plt.xlabel('Number of Comorbidities')
plt.ylabel('Count')
plt.show()
```

# 5. Gender and Health Outcome Disparities
- Examining diabetes prevalence across gender categories.
- Identifying differences in health outcomes based on gender.

```python
# Gender vs Diabetes prevalence
sns.countplot(x='gender', hue='diabetes', data=df)
plt.title('Diabetes Prevalence Across Gender Categories')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# 6. Anomaly Detection and Risk Stratification
- Detecting anomalies in health indicators.
- Creating health risk profiles using age, BMI, and HbA1c levels.

```python

# Detect anomalies using simple thresholds
# Outliers for BMI, Blood Glucose Level, and HbA1c Level
anomalies = df[(df['bmi'] > 50) | (df['blood_glucose_level'] > 300) | (df['HbA1c_level'] > 14)]
print("Anomalies detected:")
print(anomalies)

# Health risk profiles based on age, BMI, and HbA1c level
def health_risk_profile(row):
    risk_score = 0
    if row['age'] > 60:
        risk_score += 1
    if row['bmi'] > 30:
        risk_score += 1
    if row['HbA1c_level'] > 6.5:
        risk_score += 1
    return risk_score

df['risk_profile'] = df.apply(health_risk_profile, axis=1)

# Visualize risk profiles
sns.countplot(x='risk_profile', hue='diabetes', data=df)
plt.title('Diabetes Prevalence by Health Risk Profile')
plt.xlabel('Risk Profile Score')
plt.ylabel('Count')
plt.show()
```

# 7. Feature Engineering Opportunities
- Developing age groups, BMI categories, and composite risk scores.
- Generating interaction terms like age * BMI and BMI * HbA1c level.

```python

# Age group categorization
def categorize_age(age):
    if age <= 20:
        return '0-20'
    elif 21 <= age <= 40:
        return '21-40'
    elif 41 <= age <= 60:
        return '41-60'
    else:
        return '61+'

df['age_group'] = df['age'].apply(categorize_age)

# Composite risk score
df['composite_risk_score'] = df['bmi'] * df['HbA1c_level'] * df['blood_glucose_level']

# Interaction terms
df['age_bmi'] = df['age'] * df['bmi']
df['bmi_HbA1c'] = df['bmi'] * df['HbA1c_level']

# Preview new features
df.head()
```
